```python

from google import genai

def main():
    # 1) Create a client and chat session
    client = genai.Client(api_key=" ")
    chat = client.chats.create(model="gemini-2.0-flash")

    # 2) Initialize difficulty for adaptive learning (1=Easy, 5=Hard)
    difficulty_level = 1

    print("\nWelcome to the AI-Enhanced Coding Assistant!")
    print("This session focuses on Lesson 3.1: Stacks (LIFO).")
    print("We'll give you stack-related coding challenges, let you try multiple times,")
    print("and provide real-time AI feedback on your code.\n")
    print("Commands: 'skip' = move to a new challenge, 'quit'/'exit' = end session.\n")

    while True:
        # -- (A) Generate a dynamic stack-focused challenge
        challenge_prompt = (
            "You are an AI teaching assistant for a kid-friendly Python curriculum.\n"
            "Focus on Lesson 3.1: Stacks (LIFO). Generate ONE coding challenge about using stacks.\n\n"
            f"Difficulty level: {difficulty_level}\n\n"
            "If difficulty is 1, keep it very simple (e.g., push 2 items, pop 1, print the stack).\n"
            "If difficulty is higher, add complexity (multiple pushes/pops, conditionals, etc.).\n\n"
            "Format the challenge in a concise, friendly manner. Use Python list operations (append/pop) to simulate the stack."
        )

        print(f"--- Requesting a new stack challenge (Difficulty = {difficulty_level}) ---")
        challenge_stream = chat.send_message_stream(challenge_prompt)
        print("AI Challenge:", end=" ")
        for chunk in challenge_stream:
            print(chunk.text, end="")
        print("\n--- End of Challenge ---\n")

        # Multiple attempts for the same challenge
        while True:
            student_code = input("Paste your solution code (or 'skip' to get a new challenge):\n")
            if student_code.strip().lower() in ["quit", "exit"]:
                print("Thanks for using the AI Coding Assistant. Goodbye!")
                # Print entire conversation, then return
                print_full_history(chat)
                return
            if student_code.strip().lower() == "skip":
                print("Skipping this challenge. We'll move on to a new one.\n")
                break  # break out of attempts loop to get a fresh challenge

            # -- (B) Send the student's code to the AI for real-time feedback
            feedback_prompt = (
                "You are an AI teaching assistant for a kid-friendly Python curriculum.\n"
                "We're focusing on Lesson 3.1: Stacks (LIFO). The student wrote this code:\n\n"
                f"{student_code}\n\n"
                "Please:\n"
                "1. Identify any logic or syntax errors gently.\n"
                "2. Provide a short debugging tip if needed.\n"
                f"3. Suggest a small next step at difficulty level {difficulty_level}.\n\n"
                "Be friendly and use simple language appropriate for kids."
            )

            print("\n--- AI Feedback (Streaming) ---")
            feedback_response = chat.send_message_stream(feedback_prompt)
            for chunk in feedback_response:
                print(chunk.text, end="")
            print("\n--- End of Feedback ---\n")

            # (C) Ask the student if they want to try again on this challenge or move on
            retry = input("Type 'again' to try again, or 'new' to move on to the next challenge: ").strip().lower()
            if retry == "again":
                print("\nOkay, let's try again!")
                # Loop back for another attempt on the same challenge
            else:
                print("\nMoving on to a new challenge...\n")
                break  # break out of attempts loop

        # Access difficulty for the challenge
        user_assessment = input("Was the challenge 'too easy', 'too hard', or 'just right'? ")
        if user_assessment.strip().lower() == "too easy":
            difficulty_level = min(difficulty_level + 1, 5)
            print(f"Difficulty increased to {difficulty_level}.\n")
        elif user_assessment.strip().lower() == "too hard":
            difficulty_level = max(difficulty_level - 1, 1)
            print(f"Difficulty lowered to {difficulty_level}.\n")
        else:
            print("Difficulty remains the same.\n")

    print_full_history(chat)


def print_full_history(chat):
    """Utility to print the entire conversation log."""
    print("--- Full Conversation History ---")
    for msg in chat.get_history():
        print(f"role - {msg.role}:", msg.parts[0].text)


if __name__ == "__main__":
    main()
```
